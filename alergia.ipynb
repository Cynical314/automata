## This package implements the Alergia algorithm for PDFA identification in the limit

### The example below is the one extensively discussed in the seminal paper and is believed to be self explanatory.

```python
from wdfa import *
```

```python
sample_paper = ['110', '', '', '', '0', '', '00', '00', '', '', '', '10110', '', '', '100']
```

```python
G = WDFA()
```

```python
G.PPTA(sample_paper)
```

```python
G.plot()
```

```python
merge(G, 0, 1)
```

```python
G.plot()
```

```python
merge(G,0,4)
```

```python
G.plot()
```

# ....

```python
M = alergia(sample_words=sample_paper, alpha=0.8)
```

```python
M.plot()
```

```python
M.normalise()
```

```python
M.plot()
```

### The following example is one made up to show how the simple pattern (a+b+c)c* is learnt.

```python
sample_test = ['a'+'c'*i for i in range(50)] + ['b'+'c'*i for i in range(50)] + ['c'+'c'*i for i in range(50)]
```

```python
sample_test[:15]
```

```python
M = alergia(sample_words=sample_test, alpha=0.2); M.plot()
```

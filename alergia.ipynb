## This package implements the Alergia algorithm for PDFA identification in the limit

### The example below is the one extensively discussed in the seminal paper:

*Carrasco, Rafael & Oncina, Jose. (2002). Learning Stochastic Regular Grammars by Means of a State Merging Method. 10.1007/3-540-58473-0_144*. 

```python
from automatools.learning import merge, alergia, PPTA, lexicographic_sort
```

```python
sample_paper = ['110', '', '', '', '0', '', '00', '00', '', '', '', '10110', '', '', '100']
```

```python
G = PPTA(sample_paper, normalise=False)
```

```python
G.plot()
```

```python
merge(G, 0, 1)
```

```python
G.plot()
```

```python
merge(G,0,4)
```

```python
G.plot()
```

# ....

```python
M = alergia(sample_words=sample_paper, alpha=0.8, normalise=False, quotient=False)
```

```python
M.plot()
```

```python
M.normalise()
```

```python
M.plot()
```

### The following example is one made up to show how the simple pattern (a+b+c)c* is learnt.

```python
sample_test = ['a'+'c'*i for i in range(50)] + ['b'+'c'*i for i in range(50)] + ['c'+'c'*i for i in range(50)]
```

```python
lexicographic_sort(sample_test)[:15]
```

```python
M = alergia(sample_words=sample_test, alpha=0.2, normalise=False); M.plot()
```
